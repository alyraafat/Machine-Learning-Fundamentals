```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

# Perceptron

```python
class Perceptron:
    
    def __init__(self):
        self.weights = None
        self.bias = None
        self.activation = self.unit_step_func
    
    def forward_prop(self,x):
        linear_out = x @ self.weights + self.bias
        return self.activation(linear_out)
    
    def unit_step_func(self,x):
        return np.where(x>=0,1,0)

    def fit(self,x,y,lr,epochs):
        self.weights = np.zeros(shape=(x.shape[1],1))
        self.bias=0
        
        for _ in range(epochs):
            for idx,xi in enumerate(x):
                y_pred = self.forward_prop(xi)
                error = y[idx]-y_pred
                update = lr*error
                #print(self.weights.shape,xi.shape,np.expand_dims(xi,axis=1).shape)
                self.weights+=update*np.expand_dims(xi,axis=1)
                self.bias+=update
    
    def predict(self,x):
        return self.forward_prop(x)
    
    def accuracy(self,y_true, y_pred):
        accuracy = np.sum(y_true == y_pred) / len(y_true)
        return accuracy
```

# Create Dataset

```python
X, y = make_blobs(
        n_samples=150, n_features=2, centers=2, cluster_std=1.05, random_state=2
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=123
)
```

```python
p = Perceptron()
p.fit(X_train, y_train,lr=0.01, epochs=1000)
predictions = p.predict(X_test)
```

```python
acc = p.accuracy(np.expand_dims(y_test,axis=1),predictions)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.scatter(X_train[:, 0], X_train[:, 1], marker="o", c=y_train)
x0_1 = np.amin(X_train[:, 0])
x0_2 = np.amax(X_train[:, 0])

x1_1 = (-p.weights[0][0] * x0_1 - p.bias) / p.weights[1][0]
x1_2 = (-p.weights[0][0] * x0_2 - p.bias) / p.weights[1][0]

ax.plot([x0_1, x0_2], [x1_1, x1_2], "k")

ymin = np.amin(X_train[:, 1])
ymax = np.amax(X_train[:, 1])
ax.set_ylim([ymin - 3, ymax + 3])
```

# Sklearn

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
```

```python
clf = Perceptron(random_state=0)
clf.fit(X_train, y_train)
sklearn_preds = clf.predict(X_test)
sklearn_acc = accuracy_score(y_test,sklearn_preds)
sklearn_acc
```
