```python
def plot_loss(loss_history, epochs):
    epochs_array = np.arange(1, epochs + 1)  # Create an array of epoch numbers starting from 1
    plt.figure(figsize=(10, 6))
    plt.plot(epochs_array, loss_history, label='Loss')
    plt.title('Loss vs. Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_preds(x, y_true, y_pred):
    indices = np.arange(1, len(x) + 1)
    plt.figure(figsize=(10, 6))
    plt.scatter(indices, y_true, color="red", label="True Labels", marker="o")
    plt.scatter(indices, y_pred, color="blue", label="Predicted Labels", marker="x")

    plt.title("True vs Predicted Labels")
    plt.xlabel("Sample Index")
    plt.ylabel("Labels")

    plt.legend()

    plt.show()
```

```python
def one_hot_encode(y):
    categories = np.unique(y)
    one_hot = np.zeros(shape=(len(y),len(categories)))
    for i,a in enumerate(one_hot):
        a[y[i][0]] = 1 
    
    return one_hot
```

# Logistic Regression

```python
class  LogisticRegression:
    def __init__(self, activation="sigmoid"):
        self.w=None
        self.b=None
        self.activation=activation
        
    def forward_propagation(self,x):
        z = x @ self.w +self.b
        z = self.sigmoid(z) if self.activation=="sigmoid" else self.softmax(z)
        return z
    
    def softmax(self, x):
        exp_x = np.exp(x)
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)
    
    def sigmoid(self,x):
        return 1/(1+np.exp(-x))
    
    def cross_entropy_loss(self, y_true, y_pred):
        m = y_true.shape[0]
        y_true_one_hot = one_hot_encode(y_true)
        loss = -np.sum(y_true_one_hot * np.log(y_pred + 1e-15)) / m  # Adding a small value to prevent log(0)
        return loss

    def loss_fn(self,y_true,y_pred):
        if self.activation=="sigmoid":
            loss = -np.mean(y_true*np.log(y_pred)+(1-y_true)*np.log(1-y_pred))
        else:
            loss = self.cross_entropy_loss(y_true,y_pred)
        return loss
        
    def loss_prime(self,y_true,y_pred):
        n = len(y_true)
        loss_der = y_pred-y_true
        if self.activation=="sigmoid":
            loss_der /= n
        return loss_der
        
    def backward_propagation(self,x,lr,output_error):
        #print(x.T.shape,output_error.shape)
        dw = x.T @ output_error
        db = np.mean(output_error)
        self.w -= lr*dw
        self.b -= lr*db
        
    def fit(self,x,y_true,epochs,lr=0.01):
        if self.activation=="sigmoid":
            self.w = np.random.randn(x.shape[1],1)
        else:
            self.w = np.random.randn(x.shape[1],len(np.unique(y_true)))
        self.b = 0
        losses = []
        for epoch in range(1,epochs+1):
            y_pred = self.forward_propagation(x)
            loss = self.loss_fn(y_true,y_pred)
            losses.append(loss)
            dY = self.loss_prime(y_true,y_pred)
            self.backward_propagation(x,lr,dY)
            print(f"epoch: {epoch} : loss: {loss}")
        return losses
    
    def predict(self,x,threshold=0.5):
        preds = self.forward_propagation(x)
        if self.activation=="sigmoid":
            preds = preds>=threshold
        else:
            preds = np.argmax(preds,axis=1)
        return preds
            
```

```python
my_model = LogisticRegression()
my_losses = my_model.fit(X_train,y_train,100)
```

```python
plot_loss(my_losses,100)
```

```python
plot_preds(X_test,y_test,my_preds)
```

```python
my_acc = accuracy_score(y_test,my_preds)
print(f"accuracy: {my_acc}")
```
