```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn import datasets
```

```python
class LDA:
    def __init__(self,n_components):
        self.n_components = n_components
        self.linear_discriminants = None
        
    def fit(self,x,y):
        n_samples,n_features = x.shape
        class_labels = np.unique(y)
        mean_overall = np.mean(x,axis=0)
        print(mean_overall.shape)
        Sw = np.zeros((n_features,n_features)) # within class
        Sb = np.zeros((n_features,n_features)) # between classes
        for c in class_labels:
            x_c = x[y==c] # shape = n_c,n_features
            mean_c = np.mean(x_c,axis=0)
            # (n_features,n_c) . (n_c,n_features) =  (n_features,n_features)
            Sw += (x_c-mean_c).T @ (x_c-mean_c)
            
            n_c = x_c.shape[0]
            between_classes_diff = (mean_c-mean_overall).reshape(-1,1) # shape (n_features,1)
            Sb +=n_c * between_classes_diff @ between_classes_diff.T
        
        Sw_inv = np.linalg.inv(Sw)
        target_mat = Sw_inv @ Sb
        eigenvalues,eigenvectors = np.linalg.eig(target_mat)
        eigenvectors = eigenvectors.T
        idxs = np.argsort(abs(eigenvalues))[::-1]
        eigenvalues = eigenvalues[idxs]
        eigenvectors = eigenvectors[idxs]
        self.linear_discriminants = eigenvectors[:self.n_components]
        return eigenvectors, eigenvalues
    def transform(self,x):
        return x @ self.linear_discriminants.T
```

```python
data = datasets.load_iris()
X, y = data.data, data.target

# Project the data onto the 2 primary linear discriminants
lda = LDA(2)
eigenvectors,eigenvalues = lda.fit(X, y)
X_projected = lda.transform(X)

print(eigenvectors,eigenvalues)
print("Shape of X:", X.shape)
print("Shape of transformed X:", X_projected.shape)

x1, x2 = X_projected[:, 0], X_projected[:, 1]

plt.scatter(
    x1, x2, c=y, edgecolor="none", alpha=0.8, cmap=plt.cm.get_cmap("viridis", 3)
)

plt.xlabel("Linear Discriminant 1")
plt.ylabel("Linear Discriminant 2")
plt.colorbar()
plt.show()
```
