## Closed Form of Linear Regression (NHormal Equation)
I am using the pseudo inverse to account for the non-inverability of the matrix

```python
def linear_regression_closed_form_pinv(X, y):
    # Add a column of ones to X to account for the bias term (intercept)
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    
    # real normal equation
    #theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
    # Calculate the closed-form solution using the pseudo-inverse
    theta_best = np.linalg.pinv(X_b).dot(y)
    
    return theta_best

theta = linear_regression_closed_form_pinv(diabetes_X_train, diabetes_y_train)
```

## Linear Regression With Gradient Descent

```python
class LinearRegression:
    def __init__(self, reg=None, alpha=1, r=1):
        self.w = None
        self.b = None
        self.reg = reg
        self.alpha = alpha
        self.r = r
    
    def loss_fn(self,y_true,y_pred):
        loss_error = np.sum(np.square(y_true-y_pred))/float(len(y_true))
        
        regularization = 0
        if self.reg=="ridge":
            regularization = self.alpha*0.5*np.sum(np.square(self.w))
        elif self.reg=="lasso":
            regularization = self.alpha*np.sum(abs(self.w))
        elif self.reg=="elasticNet":
            l1_term = self.alpha * self.r * np.sum(abs(self.w))
            l2_term = self.alpha * (1 - self.r) * 0.5 * np.sum(np.square(self.w))
            regularization = l1_term + l2_term        
        return loss_error + regularization
    
    def loss_prime(self,y_true,y_pred):
        n = len(y_true)
        error_derivative = (-2/n)*(y_true-y_pred) 
        # print(error_derivative.shape)
        regularization_derivative = 0
        if self.reg=="ridge":
            regularization_derivative = self.alpha*self.w
        elif self.reg=="lasso":
            regularization_derivative = self.alpha*np.sign(self.w)
        elif self.reg=="elasticNet":
            l1_derivative = self.alpha * self.r * np.sign(self.w)
            l2_derivative = self.alpha * (1 - self.r) * self.w
            regularization_derivative = l1_derivative + l2_derivative
        return error_derivative , regularization_derivative
    
    def forward_propagation(self, x):
        return x @ self.w + self.b
    
    def backward_propagation(self,x,output_error,regularization_derivative,lr):
        dw = (x.T @ output_error) + regularization_derivative
        db = np.mean(output_error)
        self.w -= lr*dw
        self.b -= lr*db
        
    def fit(self,x,y,epochs, lr=1):
        # x has shape (i,j), y has shape (i,1)
        #print(x.shape,y.shape)
        self.w = np.random.randn(x.shape[1],1)
        self.b = 0
        # print(self.w.shape)
        n = len(y)
        loss_history = []
        for epoch in range(epochs):
            y_pred = self.forward_propagation(x)
            loss = self.loss_fn(y,y_pred)
            loss_history.append(loss)
            dY,regularization_derivative = self.loss_prime(y,y_pred)
            # print(dY.shape)
            self.backward_propagation(x,dY,regularization_derivative,lr)
            st= f"epoch {epoch+1}: loss: {loss}"
            print(st)
        return loss_history
    def predict(self,x):
        return self.forward_propagation(x)
            
```

```python
my_model = LinearRegression()
my_losses = my_model.fit(diabetes_X_train,diabetes_y_train,2000)
```

```python
def plot_loss(loss_history, epochs):
    epochs_array = np.arange(1, epochs + 1)  # Create an array of epoch numbers starting from 1
    plt.figure(figsize=(10, 6))
    plt.plot(epochs_array, loss_history, label='Loss')
    plt.title('Loss vs. Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_loss(my_losses,2000)
```

```python
diabetes_y_pred = my_model.predict(diabetes_X_test)
```

```python
plot_preds(diabetes_X_test,diabetes_y_test,diabetes_y_pred)
```

```python
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))
```
